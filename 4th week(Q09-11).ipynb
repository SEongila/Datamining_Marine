```python
#Q9 IQR을 이용해 이산화탄소(CO2) 이상치 수를 찾으시오-Ans
import pandas as pd
# df = pd.read_csv("air_quality.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/air_quality.csv")

# 1) IQR 계산
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
 #CO2값의 상한선 규정

outliers = df[(df['CO2'] < lower) | (df['CO2'] > upper)] #CO2값의 이상치 계산.

print(len(outliers))

#Q10-Ans.
import pandas as pd
# df = pd.read_csv("drinks.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/drinks.csv")

#1) 대륙별 맥주 소비량의 평균을 계산하고, 평균이 가장 큰 대륙을 찾으시오.
continent = df.groupby("continent")['beer_servings'].mean() #Cotinent의 맥주 소비량의 평균값 구하기
top_continent = continent.idxmax() #제일 큰값의 label 값 -> 맥주 소비량이 가장 많은 대륙이름

#2) 1번에서 찾은 대륙에서 맥주 소비량이 5번째로 많은 국가(‘country’)의 맥주 소비량을 구하시오.
cond = df['continent'] == top_continent
df = df[cond]
df = df.sort_values('beer_servings', ascending=False)
ans = df.iloc[4, 1] #or df.iloc[4]['beer_servings'] -> .iloc(,) : 0번 째 행, 0번 째 열
print(ans)

#Q11-Ans 첫 번째 방법
import pandas as pd

# df = pd.read_csv("tourist.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/tourist.csv")

#1) 주어진 데이터에서 방문객 합계 및 관광객 비율을 계산 하시오.
df['방문객합계'] = df['관광'] + df['공무'] + df['사업'] + df['기타']
df['관광객비율'] = df['관광'] / df['방문객합계']

#2) 국가2에 ‘공무’와 ‘사업’으로 다녀온 방문객 수의 합을 구하시오.
a = df.sort_values('관광객비율', ascending=False).iloc[1, 3] # or iloc[1]['사업']
b = df.sort_values('관광', ascending=False).iloc[1, 2] # or iloc[1]['공무']
print(a+b)

#Q11-Ans 두 번째 방법
#합계 및 비율은 동일 - 2)만 다르게.
a = df.nlargest(2, '관광객비율').iloc[1]['사업'] #.largest(,'') : '컬럼명' 안에서 상위 n개
b = df.nlargest(2, '관광').iloc[1]['공무']
print(a+b)

#Q12-Ans 첫 번째 방법
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# df = pd.read_csv("chem.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/chem.csv")

# 1) Min-Max 스케일링 (데이터프레임으로 만들기)
scaler = MinMaxScaler() #MinMaxScaler 객체 생성 / 최소값 -> 0 , 최대값 -> 1
df['co_scaled'] = scaler.fit_transform(df[['co']]) #co 컬럼에서 최소값, 최대값 계산(fit) -> 0~1로 transform
df['nmhc_scaled'] = scaler.fit_transform(df[['nmhc']])

#2) 표준편차 계산(Pandas 표준편차)
co_std = df['co_scaled'].std()
nmhc_std = df['nmhc_scaled'].std()
print(co_std, nmhc_std)

#3) 표준편차 차이를 구하고 소수점 3자리에서 반올림
std_diff = round(co_std - nmhc_std, 3)
print(std_diff)

##Q12-Ans 두 번째 방법

#1)Min-Max 스케일링 (transform의 결과는 넘파이)
scaler = MinMaxScaler()
co_scaled = scaler.fit_transform(df[['co']])
nmhc_scaled = scaler.fit_transform(df[['nmhc']])

#2)표준편차 계산 (넘파이 표준편차)
co_std = co_scaled.std()
nmhc_std = nmhc_scaled.std()
print(co_std, nmhc_std)

#3) 방법 1과 동일
std_diff = round(co_std - nmhc_std, 3)
print(std_diff)

#Q13 -Ans 첫 번째 방법
import pandas as pd
# df = pd.read_csv(“loan.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/loan.csv")

# 1) 총대출액 컬럼 생성
df['총대출액'] = df['신용대출'] + df['담보대출']


#2) 지역코드와 성별별로 총대출액 합계 계산
grouped = df.groupby(['지역코드', '성별'])['총대출액'].sum().unstack() #unstack() : 마지막 인덱스 레벨은 '성별'이 열로 올라감.

# 3) 성별 간 총대출액 차이가 가장 큰 지역코드 찾기 (절대값 사용)
grouped['차이'] = abs(grouped[1] - grouped[2])
result = grouped['차이'].idxmax() #대출액의 차이가 가장 큰 지역코드
print(result)

#Q13-Ans 두 번째 방법

#1) 동일

#2) 지역코드와 성별별로 총대출액 합계 계산
pivot_result = df.pivot_table(
    index='지역코드',
    columns='성별',
    values='총대출액',
    aggfunc='sum'  #같은 지역코드와 성별 조합에 여러 행이 있을 수 있으므로, 그 값들은 합계로 집계.
)

#3)
pivot_result['차이'] = abs(pivot_result[1] - pivot_result[2])
result = pivot_result['차이'].idxmax()
print(result)

#Q14-Ans 첫 번째 방법
import pandas as pd
# df = pd.read_csv("crime.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

# 1) "발생건수"와 "검거건수"를 따로 분리
cond1 = df["구분"] == "발생건수"
cond2 = df["구분"] == "검거건수"
df1 = df[cond1].iloc[:, 2:] # 발생건수만 가져오기 -> 행은 모두, 열은 2열부터 끝까지.
df2 = df[cond2].iloc[:, 2:] # 검거건수만 가져오기 ->              "

# 2) 검거율 계산 (검거건수 / 발생건수)
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df2 / df1
#reset_index(drop=True)는 필터링 후 남아 있는 행들의 인덱스를 0, 1, 2, ...처럼 다시 새로 매기는 작업. 그리고 drop=True는 기존 인덱스를 새로운 열로 남기지 않고 버린다.

# 3) 각 연도에서 검거율이 가장 높은 범죄유형 찾기
listbox = df3.idxmax(axis=1)  # #df3의 각 행에서 가장 큰 값의 열 이름

# 4) 가장 높은 검거율을 기록한 범죄유형의 검거건수 가져오기
result = 0
for index, item in enumerate(listbox):
     result = result + df2.loc[index, item] #index 번째 행, item 이라는 이름의 열에 있는 값
print(result)

#Q14-Ans 두 번째 방법
import pandas as pd
# df = pd.read_csv("crime.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

#1) 데이터를 길게 변환 -> 연도와 구분은 그대로, 나머지 범죄유형 열들을 세로로 모아서 범죄유형이라는 열과 건수라는 열로 재구성
df_long = pd.melt(df, id_vars=["연도", "구분"], var_name="범죄유형", value_name="건수") ##melt() : 데이터를 분석하기 좋은 표준 형태로 바꾸는 전처리 도구

# 2. 피벗(pivot)하여 발생건수와 검거건수를 한 행에 모으기
df_pivot = df_long.pivot_table(
    index = ["연도", "범죄유형"],
    columns = "구분",
    values="건수"
).reset_index()

# 3. 검거율 계산
df_pivot["검거율"] = df_pivot["검거건수"] / df_pivot["발생건수"]

# 4. 각 연도별로 검거율이 가장 높은 범죄유형 찾기
listbox = df_pivot.groupby("연도")["검거율"].idxmax()
max_crime = df_pivot.loc[listbox] #각 연도별로 검거율이 가장 높은 행만 df_pivot() mehtod 로 골라내서 max_crime에 저장

# 5. 각 연도별 최고 검거율 범죄유형의 검거건수를 모두 합산하기
result = max_crime["검거건수"].sum()
print(result)

#Q15-Ans 첫 번쨰 방법
import pandas as pd
# df = pd.read_csv(“hr.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")

# 1) 만족도 결측치 처리
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 2) 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움) -> #근속연수 -> 전체 평균보다 같은 부서, 같은 성과등급 사람들의 평균으로 채우는 것이 더 자연스러움.
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean") #부서와 성과등급이 같은 사람들끼리 묶어서 근속연수의 평균 구함
gm = gm.astype(int)  # 평균값을 정수로.
df["근속연수"] = df["근속연수"].fillna(gm) #평균값을 원래 각 행마다 대응되게 다시 채운 Series를 만든다.

# 3) 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
# “근속연수 1년당 연봉이 얼마나 되는가” 같은 비율값 생성 예) 연봉 = 6000, 근속연수 = 3 이면, "연봉_근속연수" 값은 2000

df_year = df.nlargest(3,'연봉_근속연수')
A = df_year.iloc[-1]["근속연수"] #-1은 마지막 행을 뜻함

#4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.nlargest(2,'연봉_만족도')
B = df_like.iloc[-1]["교육참가횟수"]

# 5) 결과 출력
result = A + B
print(result)

#Q15-Ans 두 번째 방법

# 1) 만족도 결측치 처리
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 2) 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움)
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
gm = gm.astype(int)
df["근속연수"] = df["근속연수"].fillna(gm)

# 3) 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.sort_values(by="연봉_근속연수", ascending=False)
A = df_year.iloc[2]["근속연수"]
# 3번째 행의 “근속연수”

# 4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.sort_values(by="연봉_만족도", ascending=False)
B = df_like.iloc[1]["교육참가횟수"]

# 5) 결과 출력
result = A + B
print(result)





```
